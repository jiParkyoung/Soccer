## 숫자로 된 변수를 문자열로 매칭하기

```python
football['event_type'] = football['event_type'].map({0:'소식',1:'슈팅시도',2:'코너킥',3:'파울',4:'옐로카드',5:'경고누적퇴장',6:'다이렉트퇴장',7:'교체',8:'프리킥시도',9:'오프사이드',10:'핸드볼',11:'패널티선언'})

football['event_type2'] = football['event_type2'].map({12:'키패스',13:'스루패스실패',14:'퇴장',15:'자책골'})

football['side'] = football['side'].map({1:'홈팀',2:'원정팀'})

football['shot_place'] = football['shot_place'].map({1:'대형홈런',2:'막힘',3:'좌측하단',4:'우측하단',5:'정면',6:'크게빗나감',7:'골대',8:'좌측으로빗나감',9:'우측으로빗나감',10:'홈런',11:'정면상단',12:'좌측상단',13:'우측상단'})

football['shot_outcome'] = football['shot_outcome'].map({1:'유효슈팅',2:'비유효슈팅',3:'차단',4:'골대'})
football['location'] = football['location'].map({1:'공격지역',2:'수비지역',3:'패널티박스중앙',4:'왼쪽측면',5:'오른쪽측면',6:'장거리무각',7:'좌측무각',8:'우측무각',9:'좌측패널티박스',10:'좌측패널티안쪽',11:'우측패널티박스',12:'우측패널티안쪽',13:'매우가까운',14:'패널티킥위치',15:'패널티바깥쪽',16:'중거리',17:'장거리(35yd)',18:'초장거리(40yd)',19:'기록되지않음'})

football['bodypart'] = football['bodypart'].map({1:'오른발',2:'왼발',3:'머리'})
football['assist_method'] = football['assist_method'].map({0:'없음',1:'패스',2:'크로스',3:'헤딩',4:'스루패스'})
football['situation'] = football['situation'].map({1:'필드',2:'세트피스',3:'코너킥',4:'프리킥'})

football
```

## time 칼럼을 타임스탬프로 변환하기

```python
football['datetime']=0

def to_time(min):
    return str(datetime.timedelta(minutes=min))

football['time'] = football['time'].map(lambda x: to_time(x))

football['datetime'] = football['date'] +" "+ football['time']
football
```

## 골에 대한 부분합 구하기

```python
football['cumgoal'] = football.groupby(['id_odsp', 'event_team'])['is_goal'].cumsum()

football_result = football.groupby(['id_odsp', 'event_team'], sort=False)['cumgoal'].max()
football_result = pd.DataFrame(football_result)
football_result.to_csv('football_result.csv')
football_result
```

```python
football_result=football_result.reset_index()
football_result['result']=0
for i in range(len(football_result)) :
    if i%2 == 0 : 
        if football_result.iloc[i,2] > football_result.iloc[i+1,2] :
            football_result.iloc[i,3] = '승'
        elif football_result.iloc[i,2] < football_result.iloc[i+1,2] :
            football_result.iloc[i,3] = '패' 
        else :
            football_result.iloc[i,3] = '무' 
    else : 
        if football_result.iloc[i,2] > football_result.iloc[i-1,2] :
            football_result.iloc[i,3] = '승'
        elif football_result.iloc[i,2] < football_result.iloc[i-1,2] :
            football_result.iloc[i,3] = '패' 
        else :
            football_result.iloc[i,3] = '무' 
football_result

football_result=football_result.drop(['cumgoal'],axis=1)
football_result
```

```python
football=pd.merge(football,football_result,how='left',on=['id_odsp','event_team'])
football.head()
```
